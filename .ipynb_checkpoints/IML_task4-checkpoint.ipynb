```python
# init dataset
class MyDataset(data.Dataset):
  def __init__(self, X, Y):
    self.X = X
    self.Y = Y

  def __len__(self):
    return self.X.shape[0]

  def __getitem__(self, index):
    if self.Y == None:
      return self.X[index]
    else:
      return self.X[index], self.Y[index]
```

```python
# split dataset into train and val
train_data = pretrain_features[:45000]
train_labels = pretrain_labels[:45000]

dev_data = pretrain_features[45000:]
dev_labels = pretrain_labels[45000:]
```

```python
# Training Dataset
train_dataset = MyDataset(train_data, train_labels)
train_loader = data.DataLoader(
           dataset=train_dataset, 
           batch_size=256,
           num_workers=2,
           shuffle=True,
           drop_last=True,
           pin_memory=True) 

# Validation Dataset
dev_dataset = MyDataset(dev_data, dev_labels)
dev_loader = data.DataLoader(
          dataset=dev_dataset,           
          batch_size=256,
          num_workers=2,
          shuffle=False,
          pin_memory=True)
```

```python
class FeatureExtractor(nn.Module):
  def __init__(self):
    super().__init__()

    # BN after activation
    self.layers = nn.Sequential(nn.Linear(1000,1024),
                                    nn.ReLU(),
                                    nn.BatchNorm1d(1024),
                                    nn.Dropout(p=0.2),
                                    nn.Linear(1024, 512),
                                    nn.ReLU(),
                                    nn.BatchNorm1d(512),
                                    nn.Dropout(p=0.2),
                                    nn.Linear(512, 256),
                                    nn.ReLU(),
                                    nn.BatchNorm1d(256),
                                    nn.Dropout(p=0.2))
    
  def forward(self, x):
    output = self.layers(x)
    
    return output
```

```python
class PretrainDecoder(nn.Module):
  def __init__(self):
    super().__init__()

    self.layers = nn.Linear(256, 1)
    
  def forward(self, x):
    output = self.layers(x)
    
    return output
```

```python
def train(encoder, decoder, train_loader, en_optimizer, de_optimizer, criterion):
  # set the model into training mode
  encoder.train()
  decoder.train()

  # initialize loss and total_correct
  running_loss = 0.0

  for id, (x, y) in tqdm(enumerate(train_loader), total=len(train_loader)): # use tqdm to monitor the training process

    # pdb.set_trace()

    # clear the gradient
    en_optimizer.zero_grad()
    de_optimizer.zero_grad()

    # store x and y in GPU
    x, y = x.to(device), y.to(device) 

    # print(x)

    # print(y)

    # predict the output using the current model
    x = encoder(x.float())
    y_hat = decoder(x.float()) #######

    # print(y_hat)

    # compute the loss
    # have to mannually cast due to MSELoss
    loss_value = criterion(y_hat.float(), y.float())   
    running_loss += loss_value.item()

    # backprop and update the weights
    loss_value.backward()
    en_optimizer.step()
    de_optimizer.step()
    
  avg_loss = running_loss / len(train_loader)
  return avg_loss

def eval(encoder, decoder, dev_loader, criterion):
  encoder.eval()
  decoder.eval()

  running_loss = 0.0

  with torch.no_grad():
    for id, (x, y) in tqdm(enumerate(dev_loader), total=len(dev_loader)): 

      # store x and y in GPU
      x, y = x.to(device), y.to(device) 

      # predict the output using the current model
      x = encoder(x.float())
      y_hat = decoder(x.float())

      # compute the loss
      loss_value = criterion(y_hat.float(), y.float()) 
      running_loss += loss_value.item()

  avg_loss = running_loss / len(dev_loader)
  return avg_loss
```

```python
# Initialize the model
encoder = FeatureExtractor().to(device)
decoder = PretrainDecoder().to(device)
# model = torch.load('/content/gdrive/MyDrive/11785/HW1P2/best_6_moreDropout_1.pt').to(device)
print(encoder)
print(decoder)

# Optimizer (Adam(*), SGD + Momentum, etc)
en_optimizer = torch.optim.Adam(encoder.parameters(), lr=0.001)
de_optimizer = torch.optim.Adam(decoder.parameters(), lr=0.001)

# Sheduler 
# scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, 32)

# Criterion
criterion = nn.MSELoss()

# Define the number of epochs and initialize the best accuracy
n_epochs = 10
best_loss = 10000
```

```python
n_epochs = 5
```

```python
for g in en_optimizer.param_groups:
    g['lr'] = 0.00001

for g in de_optimizer.param_groups:
    g['lr'] = 0.00001
```

```python
for epoch in range(n_epochs):
  # Training

  avg_loss = train(encoder, decoder, train_loader, en_optimizer, de_optimizer, criterion)

  # print out the average loss and accuracy
  print(f"epoch: {epoch}, train_avg_loss: {avg_loss}")

  # Evaluation
  avg_loss = eval(encoder, decoder, dev_loader, criterion)

  # print out the average loss and accuracy
  print(f"epoch: {epoch}, eval_avg_loss: {avg_loss}")

  # Save the best model
  if avg_loss < best_loss:
    print(f"save model!")
    torch.save(encoder,'best_pretrain_encoder.pt')
    best_loss = avg_loss
  
  # Call scheduler
  # scheduler.step(accuracy)
```

```python
class TransDecoder(nn.Module):
  def __init__(self):
    super().__init__()

    self.layers = nn.Sequential(nn.Linear(256, 256),
                                nn.ReLU(),
                                nn.BatchNorm1d(256),
                                nn.Linear(256, 1))
    
  def forward(self, x):
    output = self.layers(x)
    
    return output
```

```python
def trans_train(encoder, decoder, train_loader, de_optimizer, criterion):
  # set the model into training mode
  encoder.eval()
  decoder.train()

  # initialize loss and total_correct
  running_loss = 0.0

  for id, (x, y) in tqdm(enumerate(train_loader), total=len(train_loader)): # use tqdm to monitor the training process

    # pdb.set_trace()

    # clear the gradient
    # en_optimizer.zero_grad()
    de_optimizer.zero_grad()

    # store x and y in GPU
    x, y = x.to(device), y.to(device) 

    # predict the output using the current model
    with torch.no_grad():
      x = encoder(x.float())
    
    y_hat = decoder(x.float()) #######

    # compute the loss
    # have to mannually cast due to MSELoss
    loss_value = criterion(y_hat.float(), y.float())   
    running_loss += loss_value.item()

    # backprop and update the weights
    loss_value.backward()
    # en_optimizer.step()
    de_optimizer.step()
    
  avg_loss = running_loss / len(train_loader)
  return avg_loss
```
